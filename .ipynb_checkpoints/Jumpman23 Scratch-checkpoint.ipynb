```python
def convertToSeconds(timestr):
    timestr
    ftr = [3600,60,1]
    try:
        return sum([a*b for a,b in zip(ftr, map(float,timestr.split(':')))])
    except: 
        pass
    
def calculateDistance(lat1,lon1,lat2,lon2):
    start_coord = (lat1,lon1)
    end_coord = (lat2,lon2)
    return(vincenty(start_coord,end_coord).miles)
    

df['how_long_it_took_to_order'].apply(convertToSeconds)
```

# Batch Matching

```python
def batchSimulate(df, interval):
    #simulating batch matching with set interval in minutes
    tdelta = datetime.timedelta(minutes=interval)
    oldest = min(df['when_the_delivery_started'])
    youngest = max(df['when_the_delivery_started'])
    
    return (oldest, oldest + tdelta)
```

```python
batchSimulate(df3, 5)
```
